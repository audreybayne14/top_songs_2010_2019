```python
import pandas as pd

# load data and change encoding
data = pd.read_csv("top10s.csv", encoding='latin1')

# Display the first few rows of the data
data.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set up the visual style
sns.set(style="whitegrid")
```

```python
# Trend in Loudness Over the Years
plt.figure(figsize=(14, 6))
sns.lineplot(data=data, x='year', y='dB')
plt.title('Trend in Loudness Over the Years')
plt.ylabel('Loudness (dB)')
plt.xlabel('Year')
plt.show()
```

```python
# Trend in Danceability Over the Years
plt.figure(figsize=(14, 6))
sns.lineplot(data=data, x='year', y='dnce')
plt.title('Trend in Danceability Over the Years')
plt.ylabel('Danceability')
plt.xlabel('Year')
plt.show()
```

```python
# Trend in Energy Level Over the Years
plt.figure(figsize=(14, 6))
sns.lineplot(data=data, x='year', y='nrgy')
plt.title('Trend in Energy Level Over the Years')
plt.ylabel('Energy Level')
plt.xlabel('Year')
plt.show()
```

```python
# Trend in BPM Over the Years
plt.figure(figsize=(14, 6))
sns.lineplot(data=data, x='year', y='bpm')
plt.title('Trend in BPM Over the Years')
plt.ylabel('Beats Per Minute (BPM)')
plt.xlabel('Year')
plt.show()
```

```python
# Top Genres Over the Years
plt.figure(figsize=(14, 8))
genre_trend = data.groupby(['year', 'top genre']).size().unstack().fillna(0)
genre_trend.plot(kind='area', stacked=True, figsize=(14, 8), colormap='tab20')
plt.title('Top Genres Over the Years')
plt.ylabel('Number of Songs')
plt.xlabel('Year')
plt.legend(title='Top Genre', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# Genre Popularity Over Time
plt.figure(figsize=(14, 8))
genre_heatmap = data.pivot_table(index='year', columns='top genre', aggfunc='size', fill_value=0)
sns.heatmap(genre_heatmap, cmap='YlGnBu')
plt.title('Genre Popularity Over Time')
plt.ylabel('Year')
plt.xlabel('Genre')
plt.show()
```

```python
# Distribution of BPM by Genre
plt.figure(figsize=(14, 8))
sns.boxplot(data=data, x='top genre', y='bpm')
plt.title('Distribution of BPM by Genre')
plt.ylabel('BPM')
plt.xlabel('Genre')
plt.xticks(rotation=90)
plt.show()
```

```python
# Correlation Heatmap of Song Features
numerical_data = data.select_dtypes(include=['float64', 'int64'])
plt.figure(figsize=(14, 8))
corr = numerical_data.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Heatmap of Song Features')
plt.show()
```

```python
# Top 25 Artists in Top Songs List
plt.figure(figsize=(14, 8))
top_artists = data['artist'].value_counts().head(25)
sns.barplot(x=top_artists.values, y=top_artists.index, palette='viridis')
plt.title('Top 25 Artists in Top Songs List')
plt.xlabel('Number of Songs')
plt.ylabel('Artist')
plt.show()
```

```python
# Trend in Song Duration Over the Years
plt.figure(figsize=(14, 6))
sns.lineplot(data=data, x='year', y='dur')
plt.title('Trend in Song Duration Over the Years')
plt.ylabel('Duration (seconds)')
plt.xlabel('Year')
plt.show()
```

```python
# Distribution of Popularity Scores
plt.figure(figsize=(14, 6))
sns.histplot(data['pop'], bins=20, kde=True, color='purple')
plt.title('Distribution of Popularity Scores')
plt.xlabel('Popularity Score')
plt.ylabel('Frequency')
plt.show()

```

```python
# Word Cloud of Song Titles

from wordcloud import WordCloud

text = ' '.join(data['title'])
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

plt.figure(figsize=(14, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Song Titles')
plt.show()
```
